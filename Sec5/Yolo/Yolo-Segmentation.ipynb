```python
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load a model (YOLOv8)
model = YOLO(r"D:\Subject\Computer Vision\Sections\yolov8n-seg.pt")  # Make sure the path is correct

# Predict with the model (using an image URL)
results = model(r"D:\Subject\Computer Vision\Sections\cat.jpg")  # This URL points to an image of a bus

result_img = results[0].plot()
plt.imshow(result_img)
plt.axis('off')
plt.show()
```
